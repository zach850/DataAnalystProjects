Print the highest weekly sales for each "department" in the "walmart" DataFrame 

Limit your results to the top five departments, in descending order. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
walmart_df = pd.read_csv("sales_subset.csv")
walmart_df
walmart_df = walmart_df.drop(columns=['Unnamed: 0'])
walmart_df
```

```python
walmart_df
how_many_departments = walmart_df['department'].unique()
how_many_departments
```

```python
# Method 1
walmart_df.groupby('department')['weekly_sales'].max()
```

```python
# Method 2 
walmart_df.groupby('department').agg({'weekly_sales':'max'}).sort_values('weekly_sales', ascending=False).head(5)
```
