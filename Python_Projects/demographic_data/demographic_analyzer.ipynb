# What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?

```python
df4 = df[df['education-num'] >= 13].sum()

total = df[['education', 'salary']] \
    .query('salary == ">50K" & education == ["Bachelors", "Masters", "Doctorate" ]') 

percent = total['salary'].value_counts() /  df4['education-num'] *100
percent
```
