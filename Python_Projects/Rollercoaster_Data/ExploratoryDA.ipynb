```python
df = df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date',
    # 'Type', 
      'Manufacturer', 
      #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
      # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
      # 'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
      # 'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 
       #'Opened', 
       #'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
       #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
       #'height_value', 'height_unit', 
       'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.dtypes
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
df.dtypes
```

```python
# Rename columns 
df = df.rename(columns={'coaster_name': 'Coaster_Name',
                   'year_introduced' : 'Year_Introduced',
                   'latitude': 'Latitude',
                   'longitude' : 'Longitude',
                   'opening_date_clean' : 'Opening_Date',
                   'speed_mph' : 'Speed_mph', 
                   'height_ft' : 'Height_ft',
                  'Inversions_clean' : 'Inversions',
                  'Gforce_clean' : 'Gforce'})
```

```python
df.head()
```

```python
#Find null values 
df.isna().sum()
```

```python
#Find duplicated values 
df.loc[df.duplicated()]
```

```python
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])] \
    .reset_index(drop=True).copy()
```

```python
# Feature Understanding 
df['Year_Introduced'].value_counts()
```

```python
ax = df['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Top 10 Years Coasters Introduced')
ax.set_xlabel('Year_Introduced')
ax.set_ylabel('Count')
```

```python
ax = df['Speed_mph'].plot(kind='hist', bins=20, title= "Coaster Speed (mph)")
ax.set_xlabel('Speed mph')
ax.set_ylabel('Count')
```

```python
# Feature Relationships
df.plot(kind='scatter', 
        x='Speed_mph', 
        y='Height_ft',
        title= 'Coaster Speed v Height')
plt.show()
```

```python
sns.scatterplot(x='Speed_mph', 
        y='Height_ft',
        hue= 'Year_Introduced',
        data=df)
```

```python
# Ask question about the data 
# What are the locations with the fastest rollercoasters? Min of 10 coasters at that location 
df.head()
```

```python
df['Location'].value_counts()
```

```python
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean', 'count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', title="Average Coaster Speed By Location")
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
